```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
from statsmodels.iolib.table import SimpleTable, default_txt_fmt
import sklearn as sk

%matplotlib inline
```

```python
zip_dist = pd.read_excel('./Zip Codes Wake and Durham County.xlsx', sheet_name = 1, index_col = [0])

codes = np.array(zip_dist.columns)
```

```python
df = pd.read_stata("./Merged/Zestimate2.0.dta")
```

```python
codes_in_set = np.array([int(x) for x in df['zip'].unique()])
```

```python
zip_dist
```

```python
# create empty dictionary
code_dict = {}

# for each zip code, c
for c in codes:
    # create a dictionary
    cd = {}
    # for each zip code
    for code in codes:
        # distance between c and code
        cd[str(code)] = zip_dist[c][code]
    
    # for c, code_dict[c] saves all the distances of other zips from this zip
    code_dict[c] = cd
```

```python
for code in codes:
    df['distTo'+ str(code)] = df['zip'].map(code_dict[code])
```

```python
for c in df.columns:
    print(c)
```

```python
df2 = df
for index, row in df[df['lnvalue'].isna()].iterrows():
    df2.at[index, 'lnvalue'] = np.log(df2.at[index, 'salespriceamount'])
```

```python
X = np.array(df2['cum_Fee'])
# X = sm.add_constant(X)
for i in range(2,54):
    X = np.column_stack((X, np.array(df2['saleqrt'+str(i)])))
```

```python
X
```

```python
res_se = 0

for c in codes:
            model = sm.WLS(np.array(df2['lnvalue']), X, weights = 1/((1*np.array(df2['distTo'+str(c)]))**1 + 1))
            res = model.fit()
            res_se += np.average(res.bse)
print(res_se/len(codes))
```

```python
best_a = -1
best_b = -1

a_vals = np.arange(1100, 1150, 1)
b_vals = np.arange(0.38, 0.43, 0.01)

average_se = 100000

for a in a_vals:
    for b in b_vals:
        
        res_se = 0
        
        for c in codes:
            model = sm.WLS(np.array(df2['lnvalue']), X, weights = 1/((b*np.array(df2['distTo'+str(c)]))**a + 1))
            res = model.fit()
            res_se += np.average(res.bse)

        if res_se/len(codes) < average_se:
            average_se = res_se
            best_a = a
            best_b = b
```
