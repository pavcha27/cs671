```python
durham_stata = pd.read_stata('fips_37183.dta')
wake_stata = pd.read_stata('fips_37063.dta')
assessor = pd.read_csv('Assessor0/cleaned_assessor.csv')
durham_assessor = assessor[assessor['county'] == 'DURHAM'].drop(['Unnamed: 0'], axis = 1)
wake_assessor = assessor[assessor['county'] == 'WAKE'].drop(['Unnamed: 0'], axis = 1)

durham_stata
```

```python
property_list = []
for c in durham_assessor.columns:
    if 'property' in c:
        property_list.append(c)
```

```python
attempted_join = durham_stata.merge(durham_assessor, on=["property_house_number", "property_street_name", "property_city"], suffixes=['_stata','_assessor'])
```

```python
prop_list = ([c for c in property_list if c not in ["property_house_number", "property_mode", "property_street_name" , "property_city"]])
attempted_join.loc[:, ["property_house_number", "property_street_name" , "property_city"] + [c + "_stata" for c in prop_list] + [c + "_assessor" for c in prop_list]]
```
